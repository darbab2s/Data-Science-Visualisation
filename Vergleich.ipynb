### Hier ist das Ziel Ausreißer zu filtern

```python
value_u_q = daten.percentile_approx('value', percentage=25.0)
back_value_u_q = daten.percentile_approx('back_value', percentage=25.0)

#print(value_u_q)
#print(back_value_u_q)
#print()



value_m = daten.percentile_approx('value', percentage=50.0)
back_value_m = daten.percentile_approx('back_value', percentage=50.0)

#print(value_m)
#print(back_value_m)
#print()



value_o_q = daten.percentile_approx('value', percentage=75.0)
back_value_o_q = daten.percentile_approx('back_value', percentage=75.0)

#print(value_o_q)
#print(back_value_o_q)
#print()
```

```python
daten = daten.sort('value', ascending=True)
daten = daten.to_pandas_df()
```

###### Ausreißer verwerfen

```python
#print(type(daten))


value_ior = (value_o_q - value_u_q)
back_value_ior = (back_value_o_q - back_value_u_q)
#print(value_ior)
#print(back_value_ior)
#print()



un_val = value_u_q - 1.5 * value_ior
ob_val = value_o_q + 1.5 * value_ior

un_back_val = back_value_u_q - 1.5 * back_value_ior
ob_back_val = back_value_o_q + 1.5 * back_value_ior

saubere_datei = daten.loc[(daten.value >= un_val) & (daten.value <= ob_val)]
saubere_datei = saubere_datei.loc[(saubere_datei.back_value >= un_back_val) & (saubere_datei.back_value <= ob_back_val)]

#print(saubere_datei)
print(len(saubere_datei))



fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)

ax.boxplot(daten['value'], vert = 1)
ax.set_xticklabels(['value'])

plt.show()

fig = plt.figure(figsize =(10, 7))
bx = fig.add_subplot(111)

bx.boxplot(daten['back_value'], vert = 1)
bx.set_xticklabels(['back_value'])

plt.show()


fig = plt.figure(figsize =(10, 7))
ax = fig.add_subplot(111)

ax.boxplot(saubere_datei['value'], vert = 1)
ax.set_xticklabels(['value'])

plt.show()

fig = plt.figure(figsize =(10, 7))
bx = fig.add_subplot(111)

bx.boxplot(saubere_datei['back_value'], vert = 1)
bx.set_xticklabels(['back_value'])

plt.show()
```

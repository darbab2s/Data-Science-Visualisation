```python
import seaborn as sns
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def EC(a,b,x,p):
    # y^2 = x^3 +ax+b (mod p)
    return math.sqrt((x*x*x + a*x+b)%p)

def ECpow2(a,b,x,p):
    # y^2 = x^3 +ax+b (mod p)
    return ((x*x*x + a*x+b)%p)
```

```python
a = 27
print(EC(1,2,1,13))
```

```python
point = []
a = 3
b = 11
p = 13
for x in range(13):
    for y in range(13):
        if ECpow2(a,b,y,p) == (x*x %p) :
            point.append((y%p,x%p))
            point = list(set(point))
#print(point)

ECOverP13WithMinus = point+[(i[0],i[1]*-1) for i in point]
print(ECOverP13WithMinus)
```

```python
def clearEC(ec: [float])-> [float]:
    return [i for i in ec if float(int(i[0])) == i[0] and float(int(i[1])) == i[1]]
```

```python
ECOverP13 = clearEC(ECOverP13WithMinus)
print(ECOverP13)
```

```python
sns.set_theme()
fig, ax = plt.subplots(1, 1)
ax.get_xaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.grid(b=True, which='major', color='w', linewidth=1.0)
ax.grid(b=True, which='minor', color='w', linewidth=0.5)
sns.scatterplot(x=[i[0] for i in ECOverP13WithMinus],y=[i[1] for i in ECOverP13WithMinus])
```

```python
y, x = np.ogrid[-14:14:100j, -14:14:100j]
fig, ax = plt.subplots(1, 1)
ax.get_xaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.get_yaxis().set_minor_locator(mpl.ticker.AutoMinorLocator())
ax.grid(b=True, which='major', color='w', linewidth=1.0)
ax.grid(b=True, which='minor', color='w', linewidth=0.5)
plt.contour(x.ravel(), y.ravel(), (pow(y, 2) - pow(x, 3) - x * a - b), [0])
#plt.axline((-1, 0), (7, 12))
plt.grid(True)
plt.show()
```
